## Query materialized annotation tables

```python
from caveclient import CAVEclient
```

```python
url = "https://global.brain-wire-test.org/"
datastack = "h01_c3_flat"
```

```python
client = CAVEclient(datastack_name=datastack, server_address=url)
```

```python
datastack = client.info.get_datastack_info('h01_c3_flat')
datastack
```

```python
client.materialize.get_versions()
```

```python
client.materialize.version = 269
```

```python
# get all available tables
client.materialize.get_tables()
```

```python
# get head of the nucleus table
client.materialize.query_table(table="nucleus", 
                               limit=10)
```

```python
# query cells table for pyramidal cells
client.materialize.query_table(table="cells",
                                filter_equal_dict = {'cell_type': "PYRAMIDAL"}, 
                                limit=10)
```

```python
# spatial filters
# https://caveclient.readthedocs.io/en/latest/guide/materialization.html#spatial-filters
bounding_box = [[100000, 100000, 1000], [200000, 200000, 2000]]
client.materialize.query_table(table="cells",
                                filter_spatial_dict = {'pt_position': bounding_box})
```
