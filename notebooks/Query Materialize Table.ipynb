```python
import caveclient
```

```python
client = caveclient.CAVEclient('h01_c3_flat')
```

```python
datastack = client.info.get_datastack_info('h01_c3_flat')
datastack
```

```python
client.materialize.get_versions()
```

```python
version = client.materialize.version
version
```

```python
# get all available tables
client.materialize.get_tables(datastack, version)
```

```python
# get head of the nucleus table
client.materialize.query_table(table="nucleus_table", 
                               datastack_name="h01_c3_flat", 
                               limit=10)
```

```python
# query dbcells table for pyramidal cells
client.materialize.query_table(table="dbcells_dump",
                                  filter_equal_dict = {'cell_type': "PYRAMIDAL"}, 
                                  limit=10)
```
