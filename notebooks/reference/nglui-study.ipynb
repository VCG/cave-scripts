# Example 2

Example from https://github.com/seung-lab/NeuroglancerAnnotationUI/blob/master/examples/dash_with_statebuilder.py

```python
df = pd.read_hdf("../data/dash_example_data.h5", 'data')
df.head()
```

```python
image_layer = ImageLayerConfig(name='EM',
                                source='precomputed://gs://neuroglancer/pinky100_v0/son_of_alignment_v15_rechunked')
seg_layer = SegmentationLayerConfig(name='layer23',
                                    source='precomputed://gs://microns_public_datasets/pinky100_v185/seg',
                                    selected_ids_column='pt_root_id')
points = PointMapper('pt_position')
anno_layer = AnnotationLayerConfig(name='annos', mapping_rules=points)

sb = StateBuilder([image_layer, seg_layer, anno_layer])
sb.render_state(df,
                return_as='url',
                url_prefix='https://neuromancer-seung-import.appspot.com/')
```

```python
from nglui.statebuilder import *
```

```python
url = "https://global.brain-wire-test.org/"
datastack = "h01_c3_flat"
from caveclient import CAVEclient
client = CAVEclient(datastack_name=datastack, server_address=url)
```

```python
annotationsDf = client.materialize.query_table(table="cells",
#                                               desired_resolution=[4,4,40]
                                              )
```

```python
annotationsDf.tail()
```

```python
# img_source = "graphene://middleauth+https://local.brain-wire-test.org/segmentation/1.0/h01_full0_v2"
# img_source = "precomputed://gs://h01-release/data/20210601/4nm_raw"
img_source = client.info.image_source()
img_layer = ImageLayerConfig(name='h01', source=img_source)
```

```python
# seg_source = client.info.segmentation_source()
# seg_layer = SegmentationLayerConfig(name="segs", source=seg_source)
```

```python
points = PointMapper(point_column="pt_position")
anno_layer = AnnotationLayerConfig(name="annos", mapping_rules=[points])
```

```python
# This is working
sb = StateBuilder([img_layer, anno_layer], url_prefix='https://ngl.cave-explorer.org/')
d = sb.render_state(annotationsDf[100:200], 
                    return_as='dict', 
                    compatibility_mode=True, 
                    url_prefix='https://ngl.cave-explorer.org/')
```

```python
d['layers'][1]['source'][0]['transform']['outputDimensions']={
            "x": [
              4e-9,
              "m"
            ],
            "y": [
              4e-9,
              "m"
            ],
            "z": [
              3.3e-8,
              "m"
            ]
          }
```
